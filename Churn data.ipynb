```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
data = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1).copy()
data.head()
```

```python
data['Geography'].value_counts()
```

```python
data['Gender'].value_counts()
```

```python
data = data.rename(columns={'Age': 'SeniorCitizen'})
for x in range(0,10000):
    if data['SeniorCitizen'][x] < 65:
        data['SeniorCitizen'][x] = 0
    else:
        data['SeniorCitizen'][x] = 1
```

```python
df.dtypes
```

```python
for col in data:
  print(f'{col}: {data[col].unique()}')
```

```python
gender_ohe = pd.get_dummies(data["Gender"], drop_first=True)
country_ohe = pd.get_dummies(data["Geography"], drop_first=True)

# delete catfeatures
data.drop(["Gender", "Geography"], axis=1, inplace=True)

#concat new sets
data = pd.concat([data, gender_ohe, country_ohe], axis=1)

data.head()
```

```python
cols_to_scale = ['Tenure','CreditScore','EstimatedSalary','Balance']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[cols_to_scale] = scaler.fit_transform(data[cols_to_scale])
```

```python
data
```

```python
#train test split for our ANN
X = data.drop('Exited',axis='columns')
y = data['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
   keras.layers.Dense(11, input_shape=(11,)), #input layer
   keras.layers.Dense(15, activation='relu'), #hidden layer
   keras.layers.Dense(1, activation='sigmoid') #ouput layer
])

#or this will also work
# model = keras.Sequential([
#     keras.layers.Dense(32, input_shape=(32,), activation='relu'),
#     keras.layers.Dense(1, activation='sigmoid') #ouput layer
# ])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50)
```

```python
yp = model.predict(X_test)
```

```python
yp[:5]
```

```python
y_pred = []
for element in yp:
  if element > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```
